## Q13) What is the minimum and max values of each column against each weather value

```python
## to display the min values 
Weather_data.groupby('Weather').min()
```

```python
## to display the max values 
Weather_data.groupby('Weather').max()
```

```python
## here i will join the 2 results 
min_data = Weather_data.groupby('Weather').min()
max_data=Weather_data.groupby('Weather').max()
# join the 2 values 
min_max_data = min_data.join(max_data,on="Weather" , lsuffix="_min" , rsuffix="_max")
min_max_data
```
